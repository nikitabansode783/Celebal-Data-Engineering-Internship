### Missing Value Analysis
The dataset contains missing values in six columns:
- discount
- what_customers_said
- seller_name
- videos
- seller_information
- variations

Since these missing values are primarily present in optional descriptive fields, removing rows would lead to unnecessary data loss. Therefore, meaningful replacement values are used to preserve the dataset while improving data quality.

```python
#Handling Missing Values

#Filling the missing values in discount as '0'
df['discount'] = df['discount'].fillna(0)

#Filling the missing values in seller's name as  'Unknown seller'
df['seller_name'] = df['seller_name'].fillna('Unknown Seller')

#Filling the missing values in seller's information as 'Not Available'
df['seller_information'] = df['seller_information'].fillna('Not Available')

#Filling the missing values in variations as 'Not Available'
df['variations'] = df['variations'].fillna('Not Available')

#Dropping the two columns 'videos' and 'what_customers_said'
df.drop(columns=['videos','what_customers_said'], inplace=True)
```

```python
df.isnull().sum()
```

No missing values are available now. 
Therefore, we sucessfully handled the missing values

Also, The columns videos and what_customers_said contain a high percentage of missing values and are not directly relevant to the objectives of this assignment. Therefore, they were removed to simplify the dataset and focus on business-relevant attributes.

## 6. Derived Column Creation

The assignment suggests creating a derived column such as:

total_amount = price × quantity

However, after exploring the dataset, it was observed that the dataset does not contain a quantity column. Therefore, creating the exact derived feature suggested in the assignment was not possible.

An alternative feature called **Popularity Score** was created using the available attributes:

- Rating
- Ratings Count

The purpose of this feature is to identify products that are both highly rated and frequently reviewed.

Formula:

Popularity Score = Rating × Ratings Count

A product with a high rating and a large number of reviews is generally more popular and trusted by customers.

This derived feature provides meaningful business insight and helps compare products based on both quality and customer engagement.

```python
df[['rating','ratings_count']].head(10)
```

```python
df[['rating','ratings_count']].dtypes
```

```python
#Creating a new Column using 'rating' and 'rating_score'
df['popularity_score'] = df['rating'] * df['ratings_count']
df[['title', 'rating', 'ratings_count', 'popularity_score']].head(10)
```
