# 4 : Create SQLite Database

Create a temporary SQLite database and load the Superstore dataset into it. This allows us to execute SQL queries directly on the dataset.

```python
conn = sqlite3.connect("superstore.db")

df.to_sql(
    "superstore",
    conn,
    if_exists="replace",
    index=False
)

print("Database Created Successfully!")
```

# 5 : Verify Data

Display a few records from the SQL table to confirm that the dataset has been loaded successfully.

```python
query = """
SELECT *
FROM superstore
LIMIT 10;
"""

pd.read_sql(query, conn)
```

# 6 : Basic SQL Queries

Now that the dataset has been successfully loaded into the SQLite database, we can begin analyzing it using SQL queries.

### Query 1 - Total Number of Orders

```python
query = """
SELECT COUNT(*) AS Total_Orders
FROM superstore;
"""

pd.read_sql(query, conn)
```

### Query 2 - Total Sales

```python
query = """
SELECT ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore;
"""

pd.read_sql(query, conn)
```

### Query 3 - Total Profit

```python
query = """
SELECT ROUND(SUM(Profit),2) AS Total_Profit
FROM superstore;
"""

pd.read_sql(query, conn)
```

### Query 4 – Sales by Category

```python
query = """
SELECT Category,
       ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```

### Query 5 – Profit by Category

```python
query = """
SELECT Category,
       ROUND(SUM(Profit),2) AS Total_Profit
FROM superstore
GROUP BY Category
ORDER BY Total_Profit DESC;
"""

pd.read_sql(query, conn)
```

### Query 6 – Top 10 Customers by Sales

```python
query = """
SELECT `Customer Name`,
       ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY `Customer Name`
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### Query 7 – Top 10 Products by Sales

```python
query = """
SELECT `Product Name`,
       ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY `Product Name`
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### Query 8 – Sales by Region

```python
query = """
SELECT Region,
       ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY Region
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```

### Query 9 – Average Discount by Category

```python
query = """
SELECT Category,
       ROUND(AVG(Discount),2) AS Average_Discount
FROM superstore
GROUP BY Category;
"""

pd.read_sql(query, conn)
```

### Query 10 – Top 10 Most Profitable Products

```python
query = """
SELECT `Product Name`,
       ROUND(SUM(Profit),2) AS Total_Profit
FROM superstore
GROUP BY `Product Name`
ORDER BY Total_Profit DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```
